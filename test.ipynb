```python
import time
import requests
from datetime import datetime
import json
import plotly.graph_objects as go
```

```python
endpoint_url = "https://ftx.com/api/markets"
base_currency = "BTC"
quote_currency = "USD"


request_url=f"{endpoint_url}/{base_currency}/{quote_currency}"
```

```python
all_markets=requests.get(endpoint_url).json()

dh=pd.DataFrame(all_markets["result"])
dh.set_index("name",inplace=True)
dh.head(2)
```

```python
#1 dia = 60 *60 *24 segundos



daily = str (60*60*24)
#Start date = 2022-01-01
start_date = datetime(2022,1,1).timestamp()
#Get the historical market data as JSON
historical = requests.get(
    f"{request_url}/candles?resolution={daily}&start_time={start_date}"

).json()

#Convert JSON to Pandas DataFrame
dh=pd.DataFrame(historical["result"])

#Convert time to date
dh["date"] = pd.to_datetime(
    dh["time"]/1000,unit="s",origin="unix"

)
#Remove unnecessar columns
dh.drop(["startTime","time"],axis=1, inplace=True)
```

```python
dh["volume"]
```

```python
import plotly.express as px
```

```python
rollh=dh["close"].rolling(window=7).mean()
#Create a graph objet
fig=go.Figure()
#Set up the layout
fig.update_layout(
    title={
        "text":f"{base_currency}/{quote_currency}",
        "x" : 0.5,
        "xanchor":"center"
    },
    xaxis_title ="Date",
    yaxis_title = "Price",
    xaxis_rangeslider_visible=False
)

#Add candlestick chart
fig.add_trace(
    go.Candlestick(
        x=dh["date"],
        open=dh["open"],
        high=dh["high"],
        low=dh["low"],
        close=dh["close"],
        name="Candlestick(Precio)"
    )
)
fig.add_trace(
    go.Scatter(
        x=dh["date"],
        name="Media Movil(7d)",
        y=rollh
)
)

fig.show()
```
